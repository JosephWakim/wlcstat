Example usage of the 'r2_ave' function within 'wlcave' module

```python
num_pts = 100
length_kuhn_0 = 1e-5
length_kuhn_f = 1e5
length_kuhn = np.logspace(np.log10(length_kuhn_0), np.log10(length_kuhn_f), num_pts)
dimensions = 2
r2 = r2_ave(length_kuhn, dimensions)
```

```python
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)
plt.loglog(length_kuhn, r2)
plt.loglog(length_kuhn[0:60], length_kuhn[0:60] ** 2)
plt.loglog(length_kuhn[40:100], length_kuhn[40:100] * 2 / (dimensions - 1))
plt.xlabel(r'Chain length $N = L/(2l_{p})$')
plt.ylabel(r'End-to-end distance $\langle R^{2} \rangle$')
plt.tight_layout()
plt.show()
```
